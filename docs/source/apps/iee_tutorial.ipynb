# Proposed Infleqtion Tutorial Plan

## Theme

**How do we adapt AI decoders to realistic neutral-atom hardware with non-standard error mechanisms?**

The transition is `Textbook AI decoder` -> `Hardware-aware AI decoder`
---

# Learning Objectives

By the end of the Infleqtion section, participants should understand:

* Why neutral-atom systems introduce new decoding challenges.
* Why loss and leakage require additional information beyond standard syndrome decoding.
* How realistic ML training datasets are generated from quantum circuits.
* How to extend an existing neural decoder to incorporate hardware-specific information.
* How this naturally motivates current research directions in neutral-atom decoding.

---

# 40–50 minutes (10 min)

## Why Neutral Atoms are Different

Introduce the dominant non-standard error mechanisms:

* Atom loss
* Leakage
* Erasures

Compare conventional Pauli-noise assumptions with neutral-atom hardware.

Discussion points:

* Why MWPM assumes Pauli errors.
* Why atom loss violates these assumptions.
* Delayed-erasure decoding.
* Bluvstein's loss-resolved readout.
* Perrin's Leakage Detection Units (LDUs).

The goal is to motivate why hardware-aware decoding is needed.

---

# 50–65 minutes (15 min)

## Generating Realistic Training Data

Participants already know how to train a neural network.

Now demonstrate how realistic datasets are generated.

Present the [Bluvstein](https://arxiv.org/abs/2506.20661) workflow:

`Stim circuit` -> `Sample atom loss` -> `Modify circuit` ->  `Stim detector sampler` ->  `Detector events & Logical observables`


Hands-on exercise: Generate Detector events & Observable labels using Stim.

Students inspect the generated files and understand that they now possess a supervised learning dataset.
---

# 65–75 minutes (10 min)

## Extending the Neural Decoder

Start from the neural decoder trained earlier in the tutorial.

Originally:
`Detector events` -> `Neural network` ->  `Logical observable`


Now extend the input representation to include hardware information.

For example,

`Detector events & Loss information` -> `Neural network` -> `Logical observable`

Finally compare standard neural decoder, hardware-aware neural decoder and MWPM and discuss the situations where each performs well.

---

# 75–80 minutes (5 min)

## Research Outlook

Finish with the current research landscape.
